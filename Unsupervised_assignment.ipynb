## a)	Explain the Eigenvalue and eigenvector in detail along with some examples. Its role in PCA.

# answer:
### Eigenvalues and eigenvectors are fundamental concepts in linear algebra and play a crucial role in many machine learning and data analysis techniques. One of their most important applications is in Principal Component Analysis (PCA), which is used for dimensionality reduction while preserving maximum information from the data.

For a square matrix A, a non-zero vector v is called an eigenvector of A if it satisfies:

### Av = $\lambda$v
### A is a square matrix
### v is the eigenvector
### $λ$ is the corresponding eigenvalue



### Eigenvectors represent directions that remain unchanged after a linear transformation.

### Eigenvalues indicate how much the eigenvectors are stretched or compressed.

### Positive eigenvalue - stretching

### Negative eigenvalue - direction reversal

### Zero eigenvalue - collapse into a lower dimension

### example:
### Stretching a Rubber Sheet:
### Imagine a square rubber sheet that is stretched more in one direction than the other.
### The directions along which the rubber sheet stretches without changing direction are eigenvectors.
### The amount by which the sheet stretches along those directions is given by eigenvalues.

### In Principal Component Analysis (PCA), eigenvectors and eigenvalues are derived from the covariance matrix of the data.
### Eigenvectors represent the directions of the new axes, called principal components, along which the data is projected.
### Eigenvalues indicate the amount of variance captured by each corresponding eigenvector.
### Principal components are ranked based on their eigenvalues, with higher eigenvalues representing more important components.
### PCA reduces dimensionality by retaining eigenvectors with the largest eigenvalues and discarding the rest.

```python
data["genre"] = data["genre"].fillna("")
```

```python
data["type"]= data["type"].fillna(data["type"].mode()[0])
```

```python
data["rating"] = data["rating"].fillna(data["rating"].median())
```

```python
data["members_skew"] = np.log1p(data["members"])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(data['genre'])

print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

cosine_sim.shape
```

```python
indices = pd.Series(data.index, index=data['name']).drop_duplicates()
indices
```

```python
def recommend_anime(anime_title, n_recommendations=5):

    if anime_title not in indices:
        return "Anime not found in dataset"

    idx = indices[anime_title]

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:n_recommendations+1]

    anime_indices = [i[0] for i in sim_scores]

    return data[['name', 'genre', 'rating']].iloc[anime_indices]
```

```python
recommend_anime("Naruto")
```
